```python
a=open('score.csv')
a=a.read()
score_dict={}
def get_grade(a,b,c,d,e,f):
    a=float(a)
    b=float(b)
    c=float(c)
    d=float(d)
    e=float(e)
    f=float(f)
    return((a*0.25+b*0.35+c*0.4)*0.15+d*0.2+e*0.3+f*0.35)
def  get_letter(z):
    if z>=85:
        return("A")
    elif z>=70:
        return("B")
    elif z>=60:
        return("C")
    elif z>=50:
        return("D")
    else:
        return("E")
for line in a.splitlines()[1:]:
    data=line.split(',')
    studentID,Quiz1,Quiz2,Quiz3,midterm1,midterm2,final=data[0],data[1],data[2],data[3],data[4],data[5],data[6]
    total=get_grade(Quiz1,Quiz2,Quiz3,midterm1,midterm2,final)
    letter=get_letter(total)
    score_dict[studentID]=[Quiz1,Quiz2,Quiz3,midterm1,midterm2,final,total,letter]
print(score_dict)
```
